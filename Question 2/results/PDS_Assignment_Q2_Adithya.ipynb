```python
#I have Imported the required libraries
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')

df.head(10) #displayed first 10 rows
```

```python
print(df.isnull().sum()) #I have checked if there are any null values
```

##Visualization 1
I have checked the percentage of students in each gender using bar grapgh

```python
import matplotlib.pyplot as plt

student_gender_counts = df['gender'].value_counts() #counts no of students in each gender

student_gender_percentage = (student_gender_counts / student_gender_counts.sum()) * 100 # calculating its percentage

plt.figure(figsize=(6, 5))

student_gender_percentage.plot(kind='bar', color='green') #giving color for the bar graph

plt.title('Student Gender Distribution') #Shows title name

plt.xlabel('Gender')

plt.ylabel('Percentage of Students')


for index, percent in enumerate(student_gender_percentage):
    plt.text(index, percent + 1, f'{percent:.2f}%', ha='center') # for printing the percentages over the bar graphs

plt.show()
```

##Visualization 2

Checking the number of students who have completed the test preparation or not

```python
# Counting the number of students whether completed the test preparation course or not
test_prep_counts = df['test preparation course'].value_counts()

plt.figure(figsize=(5, 5))
test_prep_counts.plot(kind='bar', color='lightblue')
plt.title('Test Preparation Completion')
plt.xlabel('Completion Status')
plt.ylabel('No of Students')
plt.xticks(rotation=0)
plt.show()
```

##Visualization 3
Visualizing the percentage of students lunch distribution

```python
import matplotlib.pyplot as plt

# Counting the number of students in each lunch category
lunch_distribution = df['lunch'].value_counts()

# Creating a pie chart for lunch distribution
plt.figure(figsize=(8, 6))
lunch_distribution.plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'lightcoral'], startangle=90, wedgeprops={'edgecolor': 'black'})

plt.title('Distribution of Lunch Types')

plt.ylabel('')

plt.show()
```

##Visualization 4
Number of students in each group, based on the Parental level of education

```python
parental_education_counts = df['parental level of education'].value_counts() #Number of students in each parental level of education category

plt.figure(figsize=(10, 7))
parental_education_counts.plot(kind='bar', color='lightgreen')
plt.title('Distribution of Students Across Parental Levels of Education')
plt.xlabel('Parental Level of Education')
plt.ylabel('Number of Students')
plt.xticks(rotation=45, ha='right')

# For Adding numbers on top of each bar
for i, val in enumerate(parental_education_counts):
    plt.text(i, val + 5, str(val), ha='center', va='bottom', color='black')

plt.show()
```

##Visulaization 5

i have checked correlation between the scores, we can see that all have high scores This suggests that students who perform well in one subject tend to perform well in others.

```python
plt.figure(figsize=(6,5))
sns.heatmap(df[["math score", "reading score", "writing score"]].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Scores")
plt.show()
```
